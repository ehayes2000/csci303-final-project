```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, SimpleRNN, GRU, Embedding
from tensorflow.keras import Sequential
```

```python
# Sequential network
simple_seq_layers = [Dense(16, activation='relu'),
                     Dense(16, activation='relu'), 
                     Dense(16, activation='relu'),
                     Dense(1, activation='linear')]

simple_sequential = Sequential(simple_seq_layers) 
simple_sequential.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
simple_sequential.fit(features, target, epochs=10)
```

```python
# Sequential network
simple_auto_enc_layers = [Dense(16, activation='relu'),
                     Dense(10, activation='relu'), 
                     Dense(8, activation='relu'),
                     Dense(16, activation='relu'),
                     Dense(1, activation='linear')]

simple_sequential = Sequential(simple_auto_enc_layers) 
simple_sequential.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
simple_sequential.fit(features, target, epochs=15)
```
