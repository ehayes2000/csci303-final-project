```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, SimpleRNN, GRU, Embedding
from tensorflow.keras import Sequential
```

```python
from sklearn.preprocessing import Normalizer
normalizer = Normalizer()
features = normalizer.fit_transform(features)
```

```python
# Sequential network
simple_seq_layers = [Dense(16, activation='relu'),
                     Dense(16, activation='relu'), 
                     Dense(16, activation='relu'),
                     Dense(1, activation='linear')]

simple_sequential = Sequential(simple_seq_layers) 
simple_sequential.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
simple_sequential.fit(features, target, epochs=10)
```

```python
# Sequential network
simple_auto_enc_layers = [Dense(16, activation='relu'),
                     Dense(10, activation='relu'), 
                     Dense(8, activation='relu'),
                     Dense(4, activation='relu'),
                     Dense(1, activation='linear')]

simple_sequential = Sequential(simple_auto_enc_layers) 
simple_sequential.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
simple_sequential.fit(features, target, epochs=15)
```

```python
from tensorflow.keras.layers import Dropout
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target)

simple_auto_enc_layers = [Dense(16, activation='relu'),
                          Dropout(.1),
                          Dense(10, activation='relu'), 
                          Dropout(.1),
                          Dense(8, activation='relu'),
                          Dropout(.1),
                          Dense(4, activation='relu'),
                          Dense(1, activation='linear')]

simple_sequential = Sequential(simple_auto_enc_layers) 
simple_sequential.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
simple_sequential.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test))
```

```python
from sklearn.manifold import SpectralEmbedding
spectral_transformer = SpectralEmbedding()
spectral_embedding = spectral_transformer.fit_transform(features)
# this takes forever to run
```

```python
from tensorflow.keras.layers import Dropout
simple_auto_enc_layers = [Dense(16, activation='relu'),
                          Dropout(.1),
                          Dense(10, activation='relu'), 
                          Dropout(.1),
                          Dense(8, activation='relu'),
                          Dropout(.1),
                          Dense(4, activation='relu'),
                          Dense(1, activation='linear')]

simple_sequential = Sequential(simple_auto_enc_layers) 
simple_sequential.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
simple_sequential.fit(spectral_embedding, target, epochs=10)
```
